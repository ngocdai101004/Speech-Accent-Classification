```python
region_IDs = {
    'North' : '01',
    'North Central Coast' : '02',
    'South Central Coast' : '03',
    'Central Highland' : '04',
    'South East' : '05',
    'South West' : '06'
}
```

## Split and Change file name

```python
def split_rename_mp3_files(input_folderpath, output_folderpath, regions_data_path, out_datadump_filepath):
    
    # Ensure the directory exists
    if not os.path.isdir(input_folderpath):
        print("Directory does not exist.")
        return
    
    with open(regions_data_path, 'r') as f_regions:
        input_regions_folder = json.load(f_regions)
    
    region_count = {
        'North' : 0,
        'North Central Coast' : 0,
        'South Central Coast' : 0,
        'Central Highland' : 0,
        'South East' : 0,
        'South West' : 0
    }
    track_counter = 0
    
    with open('data_dump.json', 'r') as f_read:
        data = json.load(f_read)
    regions_layer = data["corpus"]["regions"]


    # Iterate over all folders
    for in_folder, _region in input_regions_folder.items():
        in_subfolder_path = input_folderpath + '/{}'.format(in_folder)
        region = _region['region']
        subregion = _region['subregion']
        if region_count[region] == 0:
            region_item = {
                'ID': region_IDs[region],
                'name': region,
                'subregions' : []
            }
            regions_layer.append(region_item)

        region_count[region] += 1
        out_subregion_folder = output_folderpath + '/{}/{}{:0>2}'.format(region_IDs[region], region_IDs[region], region_count[region])
        out_subregion_folder = os.path.join(out_subregion_folder)
        try:
            os.mkdir(out_subregion_folder)
        except FileExistsError:
            print(f"Directory already exists", out_subregion_folder)
            continue
        
        subregion_item = {
                'ID': '{}{:0>2}'.format(region_IDs[region], region_count[region]),
                'region': region,
                'name': subregion,
                'recordings' : []
            }
            
        segment_counter = 0
        # Iterate over all files in the directory
        for filename in os.listdir(in_subfolder_path):
            if filename.endswith(".mp3") :  # If it's an MP3 file
                mp3_file_path = in_subfolder_path + '/' + filename
                #Check existed
                if not os.path.exists(mp3_file_path):
                    continue 
                #load audio
                try:
                    audio = AudioSegment.from_mp3(mp3_file_path)
                except:
                    continue

                recording_item = {
                    "path": mp3_file_path,
                    "segments": []
                }
                # calculate the number of segments
                length = len(audio)
                length_segment = 8e3
                num_segment = len(audio) // length_segment
                if length - length_segment*num_segment > 5e3:
                    num_segment += 1
                #split audio
                for j in range(int(num_segment)): 
                    start_time = j * length_segment
                    end_time = (j + 1) * length_segment #
                    # Check for the last segment
                    if j == int(num_segment) - 1: 
                        end_time = length
                    sliced_audio = audio[start_time:end_time]
                    segment_counter += 1
                    track_counter += 1
                    output_file = out_subregion_folder + '/{}{:0>2}{:0>4}.wav'.format(region_IDs[region], region_count[region], segment_counter)
                    sliced_audio.export(output_file, format="wav")
                    segment_item = {
                        "name": '{}{:0>2}{:0>4}'.format(region_IDs[region], region_count[region], segment_counter),
						"region": region,
						"subregion": subregion,
						"start": str(start_time),
						"end": str(end_time),
						"track": str(track_counter)
                    }
                    recording_item['segments'].append(segment_item)
                subregion_item['recordings'].append(recording_item)
            elif filename.endswith(".ogg") :
                # Load the OGG file
                ogg_file_path = in_subfolder_path + '/' + filename
                try:
                    ogg_audio = AudioSegment.from_ogg(ogg_file_path)
                except:
                    continue
                segment_counter += 1
                track_counter += 1
                output_file = out_subregion_folder + '/{}{:0>2}{:0>4}.wav'.format(region_IDs[region], region_count[region], segment_counter)
                # Export the audio to WAV format
                ogg_audio.export(output_file, format='wav')
                ogg_recording_item = {
                    "path": ogg_file_path,
                    "segments": []
                }
                segment_item = {
                        "name": '{}{:0>2}{:0>4}'.format(region_IDs[region], region_count[region], segment_counter),
						"region": region,
						"subregion": subregion,
						"start": str(0),
						"end": str(len(ogg_audio)),
						"track": str(track_counter)
                    }
                ogg_recording_item['segments'].append(segment_item)
                subregion_item['recordings'].append(ogg_recording_item)
        
        for index in range(regions_layer.__len__()):
            if regions_layer[index]['ID'] == region_IDs[region]:
                regions_layer[index]['subregions'].append(subregion_item)
                break

    data["corpus"]["regions"] = regions_layer
    
    with open(out_datadump_filepath, 'w') as f_write:
        json.dump(data, f_write, indent=4)
```

```python

input_folderpath = '../dataset_raw'
output_folderpath = '../dataset'
regions_data_path = 'regions.json'
out_datadump_filepath = 'data_dump.json'

split_rename_mp3_files(input_folderpath, output_folderpath, regions_data_path, out_datadump_filepath)
```
