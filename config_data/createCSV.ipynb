```python

def createCSV(json_filepath, csv_filepath):
    with open(csv_filepath, 'w', newline='\n') as csvfile:
        header = ['filename', 'region_fold', 'subregion_fold', 'classID', 'class_name']
        csv_writer = csv.writer(csvfile)
        csv_writer.writerow(header)

        
        filename =''
        region_fold = ''
        subregion_fold = ''
        classID = ''
        class_name = ''

        with open(json_filepath, 'r') as file_datadump:
            datadump = json.load(file_datadump)
        datadump = datadump['corpus']

        for region in datadump['regions']:
            classID = region['ID']
            region_fold = region['ID']
            class_name = region['name']
            for subregion in region['subregions']:
                subregion_fold = subregion['ID']
                for recording in subregion['recordings']:
                    for segment in recording['segments'] :
                        filename = segment['name'] + '.wav'
                        csv_writer.writerow([filename, region_fold, subregion_fold, classID, class_name])


```

```python
createCSV('data_dump.json', 'VNspeech.csv')
```
