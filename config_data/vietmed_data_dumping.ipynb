```python
def create_json_file (filename):
    data = {
            'corpus': {
                'name': '',
                'subcorpus':[ {
                    'name': '',
                    'speaker-description': {
                        'name': '',
                        'gender': '',
                        'region': ''
                    },
                    'recording': {
                        'name': '',
                        'audio_path': '',
                        'segments': [
                            {
                                'name': '',
                                'region': '',
                                'start': '',
                                'end': '',
                                'track': ''
                            }
                        ]
                    }
                }
                ]
            }
        }
    # Convert the data structure to a JSON string
    json_string = json.dumps(data, indent=4)

    # Write the JSON string to a file
    with open(filename, 'w') as json_file:
        json_file.write(json_string)
   
create_json_file('data_dump.json')
```

## XML to JSON

```python
import xml.etree.ElementTree as ET
```

```python
def xml_to_json(xml_file_path, json_file_path):
    # Parse the XML file
    tree = ET.parse(xml_file_path)
    root = tree.getroot()

    # Convert XML to JSON recursively
    def xml_to_json_recursive(element):
        json_data = {}
        if element.attrib:
            json_data["@attributes"] = element.attrib
        if element.text:
            json_data["text"] = element.text.strip()
        for child in element:
            child_data = xml_to_json_recursive(child)
            if child.tag in json_data:
                if isinstance(json_data[child.tag], list):
                    json_data[child.tag].append(child_data)
                else:
                    json_data[child.tag] = [json_data[child.tag], child_data]
            else:
                json_data[child.tag] = child_data
        return json_data

    # Convert XML root element to JSON
    json_data = {root.tag: xml_to_json_recursive(root)}
    json_string = json.dumps(json_data, indent=4)
    # Write the JSON string to a file
    with open(json_file_path, 'w') as json_file:
        json_file.write(json_string)

xml_file_path = 'vietmed_corpus_v1.xml'
json_file_path = 'vietmed_corpus_v1.json'
xml_to_json(xml_file_path, json_file_path)
```

## Insert data into data_dump.json

```python
with open('vietmed_corpus_v1.json', 'r') as f_vietmed:
  vietmed = json.load(f_vietmed)

with open('data_dump.json', 'r') as f_data:
  data = json.load(f_data)

with open('regions.json', 'r') as f_rg:
  regions = json.load(f_rg)
  
```

```python
vietmed_corpus = vietmed["corpus"]
vietmed_subcorpus = vietmed_corpus["subcorpus"]
```

```python
data["corpus"]["name"] = vietmed_corpus["@attributes"]["name"]
```

```python
for subcorpus in vietmed_subcorpus:
    print (subcorpus["@attributes"]["name"])
    data_subcorpus = {
                "name": subcorpus["@attributes"]["name"],
                "speaker-description": {
                    "name": subcorpus["speaker-description"]["@attributes"]["name"],
                    "gender": subcorpus["speaker-description"]["gender"]["text"],
                    "region": regions[subcorpus["speaker-description"]["@attributes"]["name"]]
                },
                "recording": {
                    "name": subcorpus["recording"]["@attributes"]["name"],
                    "audio_path": subcorpus["recording"]["@attributes"]["audio"],
                    "segments": [
                    ]
                }
            }
    for segment in subcorpus["recording"]["segment"]:
        data_seg= {
                    "name": segment["@attributes"]["name"],
                    "region": regions[subcorpus["speaker-description"]["@attributes"]["name"]],
                    "start": segment["@attributes"]["start"],
                    "end": segment["@attributes"]["end"],
                    "track": segment["@attributes"]["track"]
                }
        data_subcorpus["recording"]["segments"].append(data_seg)
    data["corpus"]["subcorpus"].append(data_subcorpus)

with open('data_dump.json', "w") as f_data_write:
        json.dump(data, f_data_write, indent=4)
```
